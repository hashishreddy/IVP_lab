```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = 'sample.jpg'  # Change this to your image path
original = cv2.imread(image_path)
original = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
```

```python
gray = cv2.cvtColor(original, cv2.COLOR_RGB2GRAY)
```

```python
blurred = cv2.GaussianBlur(original, (15, 15), 0)
```

```python
edges = cv2.Canny(gray, 50, 150)
```

```python
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
```

```python
(h, w) = original.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 45, 1.0)
rotated = cv2.warpAffine(original, M, (w, h))
```

```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))
axes[0, 0].imshow(original)
axes[0, 0].set_title("Original Image")
axes[0, 0].axis("off")

axes[0, 1].imshow(gray, cmap='gray')
axes[0, 1].set_title("Grayscale Image")
axes[0, 1].axis("off")

axes[1, 0].imshow(blurred)
axes[1, 0].set_title("Gaussian Blurred Image")
axes[1, 0].axis("off")

axes[1, 1].imshow(edges, cmap='gray')
axes[1, 1].set_title("Canny Edge Detection")
axes[1, 1].axis("off")

axes[2, 0].imshow(thresh, cmap='gray')
axes[2, 0].set_title("Thresholded Image")
axes[2, 0].axis("off")

axes[2, 1].imshow(rotated)
axes[2, 1].set_title("Rotated Image (45°)")
axes[2, 1].axis("off")

plt.tight_layout()
plt.show()
```
