```python
image = cv2.imread('sample.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
plt.figure(figsize=(6,6))
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
hist_original = cv2.calcHist([image], [0], None, [256], [0, 256])
```

```python
equalized_image = cv2.equalizeHist(image)
```

```python
hist_equalized = cv2.calcHist([equalized_image], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(6,6))
plt.imshow(equalized_image, cmap='gray')
plt.title('Histogram Equalized Image')
plt.axis('off')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
axes[0].plot(hist_original, color='black')
axes[0].set_title('Original Histogram')
axes[1].plot(hist_equalized, color='black')
axes[1].set_title('Equalized Histogram')

plt.show()
```

```python
L = 256  # Intensity levels
modified_image = np.uint8(np.clip(image * 1.5, 0, 255))  # Increase contrast
hist_modified = cv2.calcHist([modified_image], [0], None, [256], [0, 256])
```

```python
plt.figure(figsize=(6,6))
plt.imshow(modified_image, cmap='gray')
plt.title('Modified Image (Contrast Enhanced)')
plt.axis('off')
plt.show()
```

```python
plt.figure(figsize=(6,6))
plt.plot(hist_modified, color='black')
plt.title('Modified Histogram')
plt.show()
```
